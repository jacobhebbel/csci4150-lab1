```python
import pandas as pd

df = pd.read_csv('movies.csv')
print(df.head())        # First 5 rows
print(df.columns)       # Column names
```

```python
import matplotlib.pyplot as plt

for i, row in df.iterrows():
    # if i < 10:
    #   print(f"{i} : {row['imdbRating']}, {row['WinsNoms']}, {row['IsGoodRating']}" )
    if row['IsGoodRating'] > 0:
      marker = "+"
      color = "blue"
    else:
      marker = "o"
      color = "red"

    plt.scatter(row['imdbRating'], row['WinsNoms'], marker=marker, color=color)

plt.xlabel("IMDb Rating")
plt.ylabel("Wins+Noms")
plt.title("Wins+Nominations vs. IMDb Rating")

plt.show()
```

```python
for i, row in df.iterrows():
    # if i < 10:
    #   print(f"{i} : {row['imdbRating']}, {row['Budget']}, {row['IsGoodRating']}" )
    if row['IsGoodRating'] > 0:
      marker = "+"
      color = "blue"
    else:
      marker = "o"
      color = "red"

    plt.scatter(row['imdbRating'], row['Budget'], marker=marker, color=color)

plt.xlabel("IMDb Rating")
plt.ylabel("Budget")
plt.title("Budget vs. IMDb Rating")

plt.show()
```

Wins and nominations appear to split the data better. This is because there is slightly less overlap between the two classes (Good rating and not good rating).

```python
# Part c:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# load movies.csv
movies = pd.read_csv('movies.csv')

# scatter plot of imdbVotes and imdbRating
plt.figure(figsize=(10, 6))
# Separate data by rating
good_rating = movies[movies['IsGoodRating'] == True]
not_good_rating = movies[movies['IsGoodRating'] == False]

# Plot red circles for good ratings
plt.scatter(good_rating['imdbRating'], good_rating['imdbVotes'], alpha=0.5, c='blue', marker='x', label='Good Rating', s=50)
# Plot blue crosses for not good ratings
plt.scatter(not_good_rating['imdbRating'], not_good_rating['imdbVotes'], alpha=0.5, c='red', marker='o', label='Not Good Rating', s=50)
plt.yscale('log')

plt.xlabel('IMDb Rating')
plt.ylabel('IMDb Votes')
plt.title('Votes vs. IMDb Rating Size by wins+nominations')
plt.legend()

plt.show()
```

```python
#%%
# Part d:
# scatter plot of WinsNoms vs imdbRating
plt.figure(figsize=(10, 6))
# Separate data by rating
good_rating = movies[movies['IsGoodRating'] == True]
not_good_rating = movies[movies['IsGoodRating'] == False]

# Plot red circles for good ratings
plt.scatter(good_rating['imdbRating'], good_rating['WinsNoms'], alpha=0.5, c='blue', marker='x', label='Good Rating', s=50)
# Plot blue crosses for not good ratings
plt.scatter(not_good_rating['imdbRating'], not_good_rating['WinsNoms'], alpha=0.5, c='red', marker='o', label='Not Good Rating', s=50)
plt.yscale('log')

plt.xlabel('IMDb Rating')
plt.ylabel('Wins + Nominations')
plt.title('Wins+Nominations vs. IMDb Rating Size by wins+nominations')
plt.legend()
plt.show()

# do a square root plot
plt.figure(figsize=(10, 6))
# Plot red circles for good ratings
plt.scatter(good_rating['imdbRating'], np.sqrt(good_rating['WinsNoms']), alpha=0.5, c='blue', marker='x', label='Good Rating', s=50)
# Plot blue crosses for not good ratings
plt.scatter(not_good_rating['imdbRating'], np.sqrt(not_good_rating['WinsNoms']), alpha=0.5, c='red', marker='o', label='Not Good Rating', s=50)

plt.xlabel('IMDb Rating')
plt.ylabel('Square Root of Wins + Nominations')
plt.title('Square Root of Wins+Nominations vs. IMDb Rating Size by wins+nominations')
plt.legend()
plt.show()
```

We would Want to use log scaling when data spans exponentially like the vote counts in our movie data set. This helps to compress the range of values and makes it easier to visualize patterns in the data. We would want to use square root scaling when the data is moderately skewed, as it is not aggressive as log scaling and can help to reduce the impact of outliers while preserving the structure of the data.
